觀察有興趣的欄位的資料分佈，並嘗試找出有趣的訊息

```python
import os
import numpy as np
import pandas as pd

dir_data = './data/'
```

```python
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
app_train.head()
```

```python
app_train.dtypes
```

```python
print('AMT_INCOME_TOTAL')
print('mean      | std ')
print('{0:.2f} | {1:2f}'.format(app_train['AMT_INCOME_TOTAL'].mean(), 
                                app_train['AMT_INCOME_TOTAL'].std()))
print('-------------------')
print('AMT_CREDIT')
print('mean      | std ')
print('{0:.2f} | {1:.2f}'.format(app_train['AMT_CREDIT'].mean(), 
                                app_train['AMT_CREDIT'].std()))
print('-------------------')
print('AMT_ANNUITY')
print('mean      | std ')
print('{0:.2f}  | {1:.2f}'.format(app_train['AMT_ANNUITY'].mean(), 
                                app_train['AMT_ANNUITY'].std()))
print('-------------------')
print('AMT_GOODS_PRICE')
print('mean      | std ')
print('{0:.2f} | {1:.2f}'.format(app_train['AMT_GOODS_PRICE'].mean(), 
                                app_train['AMT_GOODS_PRICE'].std()))
```

```python
sub_app_train = app_train['AMT_GOODS_PRICE'].map(lambda x: x/1000)
plt.xticks(np.linspace(sub_app_train.min(),sub_app_train.max(), 10))
plt.hist(sub_app_train, bins = 50)
```
